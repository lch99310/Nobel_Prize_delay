```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv('prize_winning_paper_record.csv')
```

```python
df.head()
```

```python
pic = sns.regplot(x="Prize_year", y="lag", data=df, order=2,
            scatter_kws={"s": 15, "alpha": 1, "color":"#CC7722"}, 
            line_kws={"color": "#960018"})

pic.set(xlabel='Year of Prize Received', ylabel='Year of Delay', title='Delay of Nobel Prize in all fields')
```

```python
fig, axs = plt.subplots(2,2, figsize=(12, 12))

pic1 = sns.scatterplot(ax=axs[0,0], data=df, x='Prize_year', y='Pub year', hue='org_country')
pic1.set(xlabel='Year of Prize Received', ylabel='Year of Publish', title='Orginaztion Country of Nobel Prize Laureates in all Fields')

pic2 = sns.scatterplot(ax=axs[0,1], data=df[df['Field'] == "Physics"], x='Prize_year', y='Pub year', hue='org_country')
pic2.set(xlabel='Year of Prize Received', ylabel='Year of Publish', title='Orginaztion Country of Nobel Prize Laureates in Physics')

pic3 = sns.scatterplot(ax=axs[1,0], data=df[df['Field'] == "Chemistry"], x='Prize_year', y='Pub year', hue='org_country')
pic3.set(xlabel='Year of Prize Received', ylabel='Year of Publish', title='Orginaztion Country of Nobel Prize Laureates in Chemistry')

pic4 = sns.scatterplot(ax=axs[1,1], data=df[df['Field'] == "Medicine"], x='Prize_year', y='Pub year', hue='org_country')
pic4.set(xlabel='Year of Prize Received', ylabel='Year of Publish', title='Orginaztion Country of Nobel Prize Laureates in Medicine')
```

可以看得出來，隨著時間越久，諾貝爾獎的延遲時間(成果發表到獲獎的時間差異)也越來越久。從二十世紀初的不到十年間就能獲獎，到二十世紀末的要超過二十年的等待才能獲獎。在2020年的近年，所頒發的獎項其研究成果多半還是來自於1990年代左右。

另外，從諾貝爾各領域的獲獎人，他們在當初研究發表所屬的組織國別可以看得出，在二十世紀初，大部分的諾獎得主都來自於歐洲，而到了二十世紀末，大部分的諾獎得主卻都來自於美國。這恰巧符合國際地緣政治在二戰前後的轉移。在1945年二戰結束前，全球是以歐洲為中心，而在1945年二戰結束以後，全球重心就轉移至美國。這也符合諾獎得主的組織國別趨勢。

總結：
1. 諾獎從成果發表到領獎，有越來越久的趨勢
2. 諾獎得主的組織所在地會隨著全球地緣中心的轉移而轉移

由以上兩的趨勢可以推斷，隨著近年中國在科研領域的不斷崛起，也在2020年在科學論文數與前1%被引用數的數量都躍升為全球第一[1]。因此我大膽預測，在2050年之後，會有越來越多的中國科學家(不光是國籍是中國，成果發表的組織也在中國)榮獲諾貝爾獎。
